```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.vectorstores import Chroma
from langchain.chains import retrieval_qa, RetrievalQA # Import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_groq import ChatGroq
import os

def initialize_llm():
  llm = ChatGroq(
    temperature=0,
    groq_api_key='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',
    model_name='llama-3.3-70b-versatile'
)
  return llm

def create_vector_db():
  # Instead of using loader_ecls inside DirectoryLoader, pass the path directly to PyPDFLoader:
  loader = PyPDFLoader("/content/mental_health_Document.pdf") # Define and initialize loader with PyPDFLoader
  document_loader = loader.load()
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
  texts = text_splitter.split_documents(document_loader)
  embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
  vector_db = Chroma.from_documents(texts, embeddings, persist_directory='./chroma_db')
  vector_db.persist()
  print("chromaDB created and data saved")

  return vector_db

def setup_qa_chain(vector_db, llm): # Define setup_qa_chain
    """Sets up the QA chain using the provided vector database and LLM."""
    qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vector_db.as_retriever())
    return qa_chain

def main():
    print("Initializing chatbot......")
    llm = initialize_llm()
    db_path = "/content/chroma_db"
    if not os.path.exists(db_path):
        vector_db = create_vector_db()
    else:
        embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
        vector_db = Chroma(persist_directory=db_path, embedding_function=embeddings)

    # Call setup_qa_chain after creating/loading the vector database
    qa_chain = setup_qa_chain(vector_db, llm)  # Assign qa_chain here

    while True:
        query = input("\nHuman: ")
        if query.lower() == "exit":
            print("chatbot: Take care of yourself, Goodbye!")
            break
        # The RetrievalQA chain's `run` method returns a string, not a dictionary.
        # So, you can directly print the response.
        response = qa_chain.run(query)
        print(f"chatbot: {response}")  # Print the response directly

    return qa_chain # Return qa_chain from main()

if __name__ == "__main__":
    qa_chain = main()  # Call main() and assign the returned qa_chain

# Now you can use qa_chain outside the main() function:
query = "your question here"  # Replace with your actual query
response = qa_chain(query) # RetrievalQA chain's `run` method returns a string, so call directly with query
print(f"chatbot: {response}")
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.document_loaders import PyPDFLoader
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_groq import ChatGroq
import os
import gradio as gr

def initialize_llm():
    llm = ChatGroq(
        temperature=0,
        groq_api_key='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',
        model_name='llama-3.3-70b-versatile'
    )
    return llm

def create_vector_db():
    loader = PyPDFLoader("/content/mental_health_Document.pdf")  # Path to the PDF file
    document_loader = loader.load()
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
    texts = text_splitter.split_documents(document_loader)
    embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
    vector_db = Chroma.from_documents(texts, embeddings, persist_directory='./chroma_db')
    vector_db.persist()
    print("chromaDB created and data saved")

    return vector_db

def setup_qa_chain(vector_db, llm):
    qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vector_db.as_retriever())
    return qa_chain

# Initialize LLM and Vector DB
print("Initializing chatbot......")
llm = initialize_llm()
db_path = "/content/chroma_db"
if not os.path.exists(db_path):
    vector_db = create_vector_db()
else:
    embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
    vector_db = Chroma(persist_directory=db_path, embedding_function=embeddings)

# Set up the QA chain
qa_chain = setup_qa_chain(vector_db, llm)

def chatbot_response(user_input, history=[]):
    if not user_input.strip():
        return "Please provide a valid input", history
    response = qa_chain.run(user_input)
    history.append((user_input, response))
    return history  # Returning history for chat interface

# Gradio chat interface setup
with gr.Blocks(theme="Respair/shiki@1.2.1") as app:
    chatbot = gr.Chatbot()  # Using Chatbot for real-time responses
    user_input = gr.Textbox(label="Your question:", placeholder="Ask something about mental health")
    gr.Button("Submit").click(chatbot_response, inputs=[user_input, chatbot], outputs=[chatbot])

    app.launch()
```
